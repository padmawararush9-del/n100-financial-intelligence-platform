```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("data/nifty100.db")

query = """
SELECT
    fr.*,
    s.broad_sector,
    mc.market_cap_crore,
    mc.pe_ratio,
    mc.pb_ratio,
    mc.dividend_yield_pct
FROM financial_ratios fr

LEFT JOIN sectors s
ON fr.company_id = s.company_id

LEFT JOIN market_cap mc
ON fr.company_id = mc.company_id
AND fr.year = mc.year
"""

df = pd.read_sql(query, conn)

print(df.columns)
df.head()
```
