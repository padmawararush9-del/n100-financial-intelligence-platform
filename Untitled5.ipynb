```python
query = """
SELECT
    fr.*,
    s.broad_sector,
    mc.market_cap_crore,
    mc.pe_ratio,
    mc.pb_ratio,
    mc.dividend_yield_pct,
    pl.sales,
    pl.net_profit,
    pl.dividend_payout
FROM financial_ratios fr

LEFT JOIN sectors s
ON fr.company_id = s.company_id

LEFT JOIN market_cap mc
ON fr.company_id = mc.company_id
AND substr(fr.year, -4) = CAST(mc.year AS TEXT)

LEFT JOIN profitandloss pl
ON fr.company_id = pl.company_id
AND fr.year = pl.year
"""

df = pd.read_sql(query, conn)
```

```python
import numpy as np
```

```python
df["cfo_pat_ratio"] = (
    df["cash_from_operations_cr"] /
    df["net_profit"]
)

df["cfo_pat_ratio"] = df["cfo_pat_ratio"].replace(
    [np.inf, -np.inf],
    np.nan
)
```

```python
df[
    [
        "cash_from_operations_cr",
        "net_profit",
        "cfo_pat_ratio"
    ]
].head()
```

```python
def normalize_score(series, reverse=False):
    """
    Winsorize (P10-P90) and normalize to 0-100.
    reverse=True is used for metrics where lower is better (e.g., Debt/Equity).
    """
    s = series.copy()

    p10 = s.quantile(0.10)
    p90 = s.quantile(0.90)

    s = s.clip(lower=p10, upper=p90)

    if reverse:
        score = (p90 - s) / (p90 - p10) * 100
    else:
        score = (s - p10) / (p90 - p10) * 100

    return score.fillna(0).clip(0, 100)
```

```python
df["roe_score"] = normalize_score(df["return_on_equity_pct"])
df["npm_score"] = normalize_score(df["net_profit_margin_pct"])
df["asset_turnover_score"] = normalize_score(df["asset_turnover"])

df["fcf_score"] = normalize_score(df["free_cash_flow_cr"])
df["cfo_pat_score"] = normalize_score(df["cfo_pat_ratio"])
df["fcf_positive_score"] = (df["free_cash_flow_cr"] > 0).astype(int) * 100

df["revenue_growth_score"] = normalize_score(df["revenue_cagr_5yr"])
df["pat_growth_score"] = normalize_score(df["pat_cagr_5yr"])

df["de_score"] = normalize_score(df["debt_to_equity"], reverse=True)
df["icr_score"] = normalize_score(df["interest_coverage"])
```

```python
df[
    [
        "roe_score",
        "fcf_score",
        "de_score",
        "revenue_growth_score"
    ]
].head()
```

```python
# Profitability (35%)
df["profitability_score"] = (
    0.15 * df["roe_score"] +
    0.10 * df["npm_score"] +
    0.10 * df["asset_turnover_score"]
)

# Cash Quality (30%)
df["cash_quality_score"] = (
    0.15 * df["fcf_score"] +
    0.10 * df["cfo_pat_score"] +
    0.05 * df["fcf_positive_score"]
)

# Growth (20%)
df["growth_score"] = (
    0.10 * df["revenue_growth_score"] +
    0.10 * df["pat_growth_score"]
)

# Leverage (15%)
df["leverage_score"] = (
    0.10 * df["de_score"] +
    0.05 * df["icr_score"]
)
```

```python
df["composite_score"] = (
    df["profitability_score"] +
    df["cash_quality_score"] +
    df["growth_score"] +
    df["leverage_score"]
).round(2)
```

```python
df[
    [
        "company_id",
        "broad_sector",
        "composite_score"
    ]
].sort_values(
    by="composite_score",
    ascending=False
).head(20)
```

```python
def sector_normalize(group):
    p10 = group.quantile(0.10)
    p90 = group.quantile(0.90)

    group = group.clip(lower=p10, upper=p90)

    return ((group - p10) / (p90 - p10) * 100).fillna(0).clip(0, 100)

df["sector_composite_score"] = (
    df.groupby("broad_sector")["composite_score"]
      .transform(sector_normalize)
      .round(2)
)
```

```python
df[
    [
        "company_id",
        "broad_sector",
        "composite_score",
        "sector_composite_score"
    ]
].sort_values(
    ["broad_sector", "sector_composite_score"],
    ascending=[True, False]
).head(20)
```
