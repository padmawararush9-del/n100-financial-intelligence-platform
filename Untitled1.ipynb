```python
import sqlite3

conn = sqlite3.connect(
    "data/nifty100.db"
)

cursor = conn.cursor()

cursor.execute(
    "SELECT name FROM sqlite_master WHERE type='table';"
)

tables = cursor.fetchall()

for table in tables:
    print(table[0])

conn.close()
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("data/nifty100.db")

tables = [
    "companies",
    "balancesheet",
    "profitandloss",
    "cashflow",
    "financial_ratios",
    "market_cap",
    "peer_groups",
    "sectors",
    "stock_prices",
    "analysis",
    "documents",
    "prosandcons"
]

for table in tables:
    count = pd.read_sql(
        f"SELECT COUNT(*) as rows FROM {table}",
        conn
    )

    print(f"{table}: {count.iloc[0,0]}")

conn.close()
```
